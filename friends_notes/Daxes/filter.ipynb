```python
img=cv2.imread("C:\\Users\\rcc\\Desktop\\man.jpg",0)
w,h=img.shape
sd=np.array([[-1,-1,-1],
             [-1,8,-1],
             [-1,-1,-1]]) 
# print(img)
for i in range(w-2):
    for j in range(h-2):
        #mat=np.array([img[i:i+3,j:j+3]])
        mat=np.array(img[i:i+3,j:j+3])
        ad=np.array([mat*sd])
        ad=np.sum(ad)
        img[i,j]=abs(ad)

# cv2.imshow("img",img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
plt.imshow(img)
```
