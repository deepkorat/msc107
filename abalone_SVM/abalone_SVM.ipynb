```python
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
df['Sex'] = l.fit_transform(df["Sex"])
```

```python
from sklearn.svm import SVC, SVR
from sklearn.model_selection import train_test_split
```

```python
X = df[['Shell weight', 'Diameter']]
y = df['Rings']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=105)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
s = SVR()
```

```python
param = {'C': [0.1,10,100,0.01], 'gamma': ['scale', 'auto'], 'kernel':['poly', 'rbf', 'linear']}
```

```python
g = GridSearchCV(s, param, cv=4)
```

```python
g.fit(X_train, y_train)
```

```python
g.best_params_
```

```python
y_pred = g.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```
